# Exploratory Data Analysis (EDA)

This section explores the processed student dataset to understand:
- Data distribution
- Relationships between features
- Impact of attendance and study hours on final marks


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## Load the Processed Dataset


```python
df = pd.read_csv(r"C:\Users\Sahana\Desktop\student-grades-project\data\processed\student_marks_processed.csv")
df.head()
```

## Dataset Overview

This shows the number of rows, columns, and data types.


```python
df.info()
```

## Statistical Summary

Provides mean, min, max, and quartile values of numeric features.


```python
df.describe()
```

## Distribution of Final Marks

This plot shows how student final marks are distributed.


```python
plt.figure(figsize=(8,5))
plt.hist(df['Final_Marks'], bins=10)
plt.xlabel("Final Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Marks")
plt.show()
```

## Attendance Percentage vs Final Marks

Analyzes how attendance affects student performance.


```python
plt.figure(figsize=(8,5))
plt.scatter(df['Attendance_Percent'], df['Final_Marks'])
plt.xlabel("Attendance Percentage")
plt.ylabel("Final Marks")
plt.title("Attendance vs Final Marks")
plt.show()
```

## Total Study Hours vs Final Marks

Shows the relationship between study hours and marks.


```python
plt.figure(figsize=(8,5))
plt.scatter(df['Total_Study_Hours'], df['Final_Marks'])
plt.xlabel("Total Study Hours")
plt.ylabel("Final Marks")
plt.title("Study Hours vs Final Marks")
plt.show()
```

## Presence Count vs Final Marks

Students with higher presence are expected to score better.


```python
plt.figure(figsize=(8,5))
plt.scatter(df['Presence_Count'], df['Final_Marks'])
plt.xlabel("Presence Count")
plt.ylabel("Final Marks")
plt.title("Presence vs Final Marks")
plt.show()
```

## Correlation Heatmap

This heatmap shows the correlation between all numerical features.


```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

## EDA Observations

- Attendance Percentage shows a positive correlation with Final Marks
- Total Study Hours moderately affect performance
- Presence count is an important contributing factor
- Dataset shows linear patterns suitable for Linear Regression


## Conclusion

EDA reveals that student engagement metrics such as attendance,
presence, and study hours significantly influence final marks.
These insights justify the use of regression models for prediction.
